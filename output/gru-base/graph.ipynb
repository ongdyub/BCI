```python
import os
import pickle

os.getcwd()
with open(file="./gru/cer.pkl", mode="rb") as f:
    gru = pickle.load(f)

with open(file="./gru-bi/cer.pkl", mode="rb") as f:
    gru_bi = pickle.load(f)

with open(file="./gru-4layer-0.01/cer.pkl", mode="rb") as f:
    gru_4 = pickle.load(f)
```

```python
with open(file="./gru/loss.pkl", mode="rb") as f:
    gru = pickle.load(f)

with open(file="./gru-bi/loss.pkl", mode="rb") as f:
    gru_bi = pickle.load(f)

with open(file="./gru-4layer-0.01/loss.pkl", mode="rb") as f:
    gru_4 = pickle.load(f)
```

```python
import matplotlib.pyplot as plt

# 그래프 그리기
plt.plot(gru[::20], label="Loss 1", color="red")
plt.plot(gru_bi[::20], label="Loss 2", color="blue")
plt.plot(gru_4[::20], label="Loss 3", color="green")

# 그래프 제목과 범례 추가
plt.title("Loss Curves")
plt.legend()

# x축 및 y축 레이블 추가
plt.xlabel("Epochs")
plt.ylabel("Loss")

# 그래프 표시
plt.show()
```
